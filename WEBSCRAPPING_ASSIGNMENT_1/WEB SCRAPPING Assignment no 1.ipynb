# 2. Write a python program to display IMDB’s Top rated 50 movies’ data (i.e. Name, IMDB rating, Year of release)and make a data frame.

```python
url='https://www.imdb.com/chart/top'
page=requests.get(url)                                       # Download IMDB's Top 50 rated movies
page
```

```python
soup=BeautifulSoup(page.content,"html.parser")               # Display page content
print(soup.prettify())
```

```python
Topresults=soup.find_all("td", class_='titleColumn')             # Scraped Top movies name
movies = []

for i in Topresults:                                             # Parse the data
    i = i.get_text().replace('\n',"")
    i=i.strip(" ")
    movies.append(i)
movies
```

```python
Scrape_ratings=soup.find_all("td",class_="ratingColumn imdbRating")         #Scrap rating movies
rating=[]

for toprated in Scrape_ratings:                                             #Parse movie rating
    toprated=toprated.get_text().replace('\n','')
    rating.append(toprated)

rating
```

```python
Scrape_releaseyear=soup.find_all('span',class_="secondaryInfo")       #Scrape the Year of Release
Released_year=[]

for year in Scrape_releaseyear:                                       #parse released year from url
    year= year.get_text().replace("\n"," ")
    Released_year.append(year)
    
Released_year
```

```python
data=pd.DataFrame({})                                       #store the scraped data
data['IMDB Movie Names']=movies
data['IMDB Ratings']=rating
data['Year_of_Release']=Released_year
data.head(50)
```

# 5.  Write a python program to scrape cricket rankings from 'www.icc-cricket.com'. You have to scrape and make data frame.
a) Top 10 ODI teams in men’s cricket along with the records for matches,points and rating.
b) Top 10 ODI Batsmen along with the records of their team and rating.
c) Top 10 ODI bowlers along with the records of their team and rating

(i) TOP 10 ODI TEAMS IN MEN CRICKET ALONG WITH RECORDS FOR MATCHES,POINTS AND RATING.

```python
def Top_10_teams(url):
    
    team=[]
    matches=[]
    points=[]
    ratings=[]
    
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    
    odi_team=soup.find_all("span",class_="u-hide-phablet")
    
    for i in odi_team:
        team.append(i.text)
    team=team[:10]
    
    #As observed from inspect,table top row style is different and other row style are different.
    #There by ,taking both top row and other rows seperately.
    
    #Match row1:-
    
    Match_top_row=soup.find_all('td',class_='rankings-block__banner--matches')
    
    for i in Match_top_row:
        matches.append(i.text)
        
        
    #for other rows:-
    other_row=soup.find_all('td',class_="table-body__cell u-center-text")
    
    for i in range(0,len(other_row),2):
        matches.append(other_row[i].text)
    matches=matches[:10]
    
    
    
    #for Point also top and other different formats
    odi_point1=soup.find_all('td',class_="rankings-block__banner--points")
    
    for i in odi_point1:
        points.append(i.text)
        
    
    odi_point2=soup.find_all('td',class_="table-body__cell u-center-text") 
        
    for i in range(1,len(other_row),2):
        points.append(odi_point2[i].text)
    points=points[:10]
        
    #Similarly for Rating:-
    
    odi_rating=soup.find_all('td',class_="rankings-block__banner--rating u-text-right")
    
    for i in odi_rating:
        ratings.append(i.text.replace('\n','').replace(' ',''))
    
    #Remaining 9 rows:--
    odi_rating=soup.find_all('td',class_="table-body__cell u-text-right rating")
    for i in odi_rating:
        ratings.append(i.text.replace('\n','').replace(' ',''))
    ratings=ratings[:10]   
    
    
    team_odi=pd.DataFrame({})
    team_odi["ODI TEAMS"]=team
    team_odi["ODI MATCHES"]=matches
    team_odi["POINTS"]=points
    team_odi["RATING"]=ratings
    
    return(team_odi)
```

```python
ODI_team_men=Top_10_teams('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
ODI_team_men
```

(ii) TOP 10 ODI BATSMEN ALONG WITH THE RECORDS OF THEIR TEAM AND RATING.

```python
def ODI_batsmen(url):
    players=[]
    teams=[]
    ratings=[]
    
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    
    
    # First row of the table
    odi_player=soup.find_all('div',class_="rankings-block__banner--name-large")
    for i in odi_player:
        players.append(i.text)
        
    #for other rows:-
    odi_player=soup.find_all('td',class_="table-body__cell rankings-table__name name")
    for i in odi_player:
        players.append(i.find('a').text)
    players=players[:10]
    
    # Firt row of teams:
    
    odi_team=soup.find_all('div',class_="rankings-block__banner--nationality")
    for i in odi_team:
        teams.append(i.text.replace('\n',''))
        
    # Other Row of teams:
    
    odi_team=soup.find_all('span',class_="table-body__logo-text")
    for i in odi_team:
        teams.append(i.text.replace('\n',''))
    teams=teams[:10]
    
    
    #First row in ratings:-
    odi_rating=soup.find_all('div',class_="rankings-block__banner--rating")
    for i in odi_rating:
        ratings.append(i.text)
        
    #other row:
    odi_rating=soup.find_all('td',class_="table-body__cell rating")
    for i in odi_rating:
        ratings.append(i.text)
    ratings=ratings[:10]
    
    batsmen_odi=pd.DataFrame({})
    batsmen_odi["PLAYERS"]=players
    batsmen_odi["TEAMS"]=teams
    batsmen_odi["RATINGS"]=ratings
    
    return(batsmen_odi)
        
    
```

```python
Top_odi_Batsmen=ODI_batsmen("https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting")
Top_odi_Batsmen
```

(iii)TOP 10 ODI BOWLERS ALONG WITH THE RECORDS OF THEIR TEAM AND RATINGS.

```python
#As batsmen and bowler page are similar in structure,thereby, using same function as above.

Top_odi_Bowlers=ODI_batsmen('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling')
Top_odi_Bowlers
```

```python
def Top_10_womenteams(url):
    
    team=[]
    matches=[]
    points=[]
    ratings=[]
    
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'html.parser')
    
    women_team=soup.find_all("span",class_="u-hide-phablet")
    
    for i in women_team:
        team.append(i.text)
    team=team[:10]
    
    #As observed from inspect,table top row style is different and other row style are different.
    #There by ,taking both top row and other rows seperately.
    
    #Match row1:-
    
    top_row=soup.find_all('td',class_='rankings-block__banner--matches')
    
    for i in top_row:
        matches.append(i.text)
        
        
    #for other rows:-
    other_row=soup.find_all('td',class_="table-body__cell u-center-text")
    
    for i in range(0,len(other_row),2):
        matches.append(other_row[i].text)
    matches=matches[:10]
    
    
    
    #for Point also top and other different formats
    odi_point1=soup.find_all('td',class_="rankings-block__banner--points")
    
    for i in odi_point1:
        points.append(i.text)
        
    
    odi_point2=soup.find_all('td',class_="table-body__cell u-center-text") 
        
    for i in range(1,len(other_row),2):
        points.append(odi_point2[i].text)
    points=points[:10]
        
    #Similarly for Rating:-
    
    odi_rating=soup.find_all('td',class_="rankings-block__banner--rating u-text-right")
    
    for i in odi_rating:
        ratings.append(i.text.replace('\n','').replace(' ',''))
    
    #Remaining 9 rows:--
    odi_rating=soup.find_all('td',class_="table-body__cell u-text-right rating")
    for i in odi_rating:
        ratings.append(i.text.replace('\n','').replace(' ',''))
    ratings=ratings[:10]   
    
    
    women_odi=pd.DataFrame({})
    women_odi["ODI TEAMS"]=team
    women_odi["ODI MATCHES"]=matches
    women_odi["POINTS"]=points
    women_odi["RATING"]=ratings
    
    return(women_odi)
```

# 7. Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame.

i) Headline
ii) Time
iii) News LinK

```python
page=requests.get("https://www.cnbc.com/world/?region=world")
soup=BeautifulSoup(page.content,'html.parser')
```

```python
#FOR HEADLINE:-
Head=[]
for i in soup.find_all("a",class_="LatestNews-headline"):
    Head.append(i.text)
   
       
```

```python
# For TIME:-
Time=[]
for i in soup.find_all("span",class_="LatestNews-wrapper"):
    Time.append(i.text)
```

```python
# FOR NEWS_LINK--
links=[]
for i in soup.find_all("div",class_="LatestNews-container"):
    links.append(i.find('a',class_="LatestNews-headline").get('href'))
print(len(links))
    
```

```python
News_website=pd.DataFrame({'HEADING':Head,'TIME':Time,'NEWS LINK':links})
News_website
```

# 9.  Write a python program to scrape mentioned details from www.dineout.co.in and make data frame.
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
Dine=requests.get("https://www.dineout.co.in/bangalore-restaurants/north-indian-cuisine")
page=BeautifulSoup(Dine.content)
page
```

```python
names=[]
for name in page.find_all("a",class_="restnt-name ellipsis"):
    names.append(name.text)
```

```python
cuisine=[]
for name in page.find_all("span",class_="double-line-ellipsis"):
    cuisine.append(name.text.split('|')[1])
```

```python
loc=[]
for name in page.find_all("div",class_="restnt-loc ellipsis"):
    loc.append(name.text)
```

```python
rating=[]
for name in page.find_all("div",class_="restnt-rating rating-4"):
    rating.append(name.text)
```

```python
image=[]
for name in page.find_all("img",class_="no-img"):
    image.append(name['data-src'])
```

```python

dineout = pd.DataFrame.from_dict({'Restaurant Name':names, 'Cuisine':cuisine, 'Location':loc, 'Rating':rating, 'Image':image},orient='index')
dineout=dineout.transpose()
dineout
```
