# Q1. Scrape the details of most viewed videos on YouTube from Wikipedia.

Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

#### Importing necessary libraries

```python
import selenium                      #importing libraries that is used to work with selenium.
from selenium import webdriver       #importing webdriver module from slenium to open automated chrome window .
from selenium.webdriver.common.by import By    # importing inbulit class by.

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from selenium.common.exceptions import NoSuchElementException 
from selenium.common.exceptions import StaleElementReferenceException

from bs4 import BeautifulSoup
import requests

import pandas as pd     # to create a dataframe 
import time

import warnings         # Importing any sort of warnings.
warnings.filterwarnings('ignore')
```

#### 1) Getting the webpage Url( https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos )

```python
Url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
```

```python
driver=webdriver.Chrome(r"C:\Users\nehas\chromedriver.exe") # Connect to the driver.
driver.maximize_window()
driver.get(Url)
time.sleep(2)
```

#### 2) Make a Empty List of all the data to be scraped.

```python
rank         = []
name         = []
artist       = []
upload_date  = []
views        = []
```

#### 3) Start scraping data using the given URL.

```python
# Extracting Rank Via X path
try:
    rnk=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[1]")
    for i in rnk:
        rank.append(i.text)
except NoSuchElementException as e:
    rank.append("EXCEPTION RAISED")
except StaleElementReferenceException:
    rank.append("EXCEPTION RAISED")
    
 # Extracting Name Via X path
try:
    nme=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[2]")
    for i in nme:
        name.append(i.text)
except NoSuchElementException as e:
    name.append("EXCEPTION RAISED")
except StaleElementReferenceException:
    name.append("EXCEPTION RAISED")
    
# Extracting Artist Name Via Xpath
try:
    artst=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[3]")
    for i in artst:
        artist.append(i.text)
except NoSuchElementException as e:
    artist.append("EXCEPTION RAISED")
except StaleElementReferenceException:
    artist.append("EXCEPTION RAISED") 
    
# Extracting Upload date Via Xpath
try:
    upload=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[5]")
    for i in upload:
        upload_date.append(i.text)
except NoSuchElementException as e:
    upload_date.append("EXCEPTION RAISED")
except StaleElementReferenceException:
    upload_date.append("EXCEPTION RAISED") 
    
# Extracting Views via Xpath
try:
    viw=driver.find_elements(By.XPATH,"//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[4]")
    for i in viw:
        views.append(i.text)
except NoSuchElementException as e:
    views.append("EXCEPTION RAISED")
except StaleElementReferenceException:
    views.append("EXCEPTION RAISED") 
```

#### 4)Check the length of each:

```python
len(rank),len(name),len(upload_date),len(artist),len(views)
```

#### 5) Create a Dataframe for scrap data

```python
Wiki_YTvideos=pd.DataFrame({'Rank':rank,'Video Name':name,'Uploader':artist,'Views (in Billons)':views,'Upload Date':upload_date})

print('\033[1m'+'Most Viewed Video on YouTube from Wikipedia :'+'\033[0m')

Wiki_YTvideos[:30]
```

```python
driver.close()
```

```python
# Extracting name via Xpath
try:
    nme_tag=driver.find_elements(By.XPATH,'//a[@class="ellipsis" ]/span')
    for i in nme_tag:
        Name.append(i.text)
except NoSuchElementException as e:
    Name.append('_')
    
    
# Extracting desgination via Xpath--
try:
    desgination_tag=driver.find_elements(By.XPATH,'//span[@class="ellipsis clr"]')
    for i in desgination_tag:
        designation.append(i.text)
except NoSuchElementException as e:
    designation.append('_')
    
    
    
# Extracting Company via Xpath--
try:
    compny_tag=driver.find_elements(By.XPATH,'//p[@class="highlightable"]/a[2]')
    for i in compny_tag:
        company.append(i.text)
except NoSuchElementException as e:
    company.append('_')
    
        
# Extracting Skill via Xpath--
try:
    skl_tag=driver.find_elements(By.XPATH,'//div[@class="hireSec highlightable"]')
    for i in skl_tag:
        skills.append(i.text)
except NoSuchElementException as e:
    skills.append('_')
    
# Extracting Location via Xpath--
try:
    loc_tag=driver.find_elements(By.XPATH,'//p[@class="highlightable"]/span[2]/small')
   
    for i in loc_tag:
        location.append(i.text)
except NoSuchElementException as e:
    location.append('_')
   
```
