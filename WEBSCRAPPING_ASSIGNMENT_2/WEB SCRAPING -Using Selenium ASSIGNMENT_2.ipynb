```python
import selenium                                                          # Import all the libraries.
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome(r"C:\Users\nehas\chromedriver.exe")             # Connect to the driver.
```

```python
url=driver.get('https://www.naukri.com/')                               # Opening Naukri.com page on automated browser.
```

```python
designations=driver.find_element(By.CLASS_NAME,"suggestor-input ")      # Entering Designation as required.
designations.send_keys('Data Analyst')
```

```python
Location=driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")#Entering Location as required.
Location.send_keys('Bangalore')
```

```python
search=driver.find_element(By.CLASS_NAME,"qsbSubmit")                   # Entering search click as required.
search.click()
```

```python
# Scraping Job Title from the given URL:-

jobtitle=[]
title_tags=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in title_tags[0:10]:
    title=i.text
    jobtitle.append(title)

    
# Scraping Job Location from the given URL:- 

joblocation=[]
loc=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for i in loc[0:10]:
    location=i.text
    joblocation.append(location)
    
# Scraping Company Name from the given URL:- 

CompanyName=[]
company=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in company[0:10]:
    com=i.text
    CompanyName.append(com)

# Scraping Experience from the given URL:- 

experiencerequired=[]
exp=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for i in exp[0:10]:
    experience=i.text
    experiencerequired.append(experience)
```

```python
# As we need to scrape first 10 Job data -->Check the length:-

print(len(jobtitle),len(joblocation),len(CompanyName),len(experiencerequired))
```

```python
# Create DataFrame:-

naukri=pd.DataFrame({'Job-Title':jobtitle,'Job Location':joblocation,'Company-Name':CompanyName,'Experience':experiencerequired})
print('\t'" Data-Analyst Jobs"'\t')
print("*"*100)
naukri
```
