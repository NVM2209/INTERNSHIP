# '''2. In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv.In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.   '''

#### 1) Fetching URL of the product in first 3 pages .

```python
URL   =[]
start =0
end   =3

for page in range(start,end):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        URL.append(i.get_attribute('href'))

    
# locating next page button and clicking
next_button=driver.find_elements(By.XPATH,'/html/body/div[1]/div[2]/div[1]/div[1]/div/span[1]/div[1]/div[70]/div/div/span/a[1]')
    
```

```python
print(len(URL))
```

#### 2) Make a Empty List of all the data to be scraped.

```python
# Make Empty list for the data to be scraped.

brand    =[]
product  =[]
price    =[]
Rx       =[]
Deli     =[]
avail    =[]
```

#### 3) Start scraping data using URL.

```python
for i in URL:
    driver.get(i)
   
    
# Extracting Brand Name via Xpath
    try:
        name=driver.find_element(By.XPATH,'//a[@id="bylineInfo" ]')
        brand.append(name.text)
    except NoSuchElementException as e:
        brand.append('-')
        
# Extracting product-Name via Xpath   
    try:
        pro=driver.find_element(By.ID, "productTitle")
        product.append(pro.text)
    except NoSuchElementException as e:
        product.append('-')
        
# Extracting Price via Xpath
    try:
        pri=driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')
        price.append(pri.text)
    except NoSuchElementException as e:
        price.append('-')
        
# Extracting Return/exchange item via Xpath 
    try:
        x=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[4]/div[4]/div[22]/div[2]/div/div/div/div[3]/span/div[2]/a')
        Rx.append(x.text)
    except NoSuchElementException as e:
        Rx.append('-')
        
# Extracting Expected Delivery via Xpath
    try:
        d=driver.find_element(By.XPATH,'//span[@class="a-text-bold"]')
        Deli.append(d.text)
    except NoSuchElementException as e:
        Deli.append('-')
 

 # Extracting Availability via Xpath
    try:
        availibility=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[4]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[3]/div/div[4]/div/div[1]/span')
        avail.append(availibility.text)
    except NoSuchElementException as e:
        avail.append('-')
    
```

```python
#Checking the length of each items:-

print(len(brand),len(product),len(price),len(Rx),len(Deli),len(avail),len(URL))
```

#### 4) Creating a dataframe of scraped data.

```python
Guitar=pd.DataFrame({'Brand':brand,'Product':product,'Price':price,'Return/Exchange':Rx,'ExpectedDelivery':Deli,'Availability':avail,'URL':URL})
pd.set_option('display.max_rows', None)
Guitar
```

```python
Guitar.to_csv('Guitar.csv')
```

```python
hostel=[]
distance=[]
rating=[]
reviews=[]
ovrrevw=[]
pvt=[]
dorm=[]
fac=[]
des=[]
url=[]
```

```python
url_tag = driver.find_elements(By.XPATH,'/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[1]/header/div[2]/div/div[2]/ul[2]/li[3]/a')
for u in url_tag:
    url.append(u.get_attribute('href'))
```

```python
len(url)
```

```python
name = driver.find_elements(By.XPATH,"//h2[@class='title title-6']//a")
for hosteln in name:
    try:
        hostel.append(hosteln.text)
    except NoSuchElementException:    
        hostel.append("-")
    
dist= driver.find_elements(By.XPATH, "//span[@class='description']/span")
for dis in dist:
    try:
        distance.append(dis.text)
    except NoSuchElementException:
        distance.append("-")

rate=driver.find_elements(By.XPATH, "//div[@class='score orange big']")
for rat in rate:
    try:
        rating.append(rat.text)
    except NoSuchElementException:    
        rating.append("-")

rev=driver.find_elements(By.XPATH, "//div[@class='reviews']")
for treview in rev:
    try:
        reviews.append(treview.text)
    except NoSuchElementException:    
        reviews.append("-")
            
            
overvw=driver.find_elements(By.XPATH, "//div[@class='keyword']/span")       
for over in overvw:
        try:
            ovrrevw.append(over.text)
        except NoSuchElementException:    
            ovrrevw.append("-")
            
            
pvtt=driver.find_elements(By.XPATH, "//div[@class='prices-col']//a[1]/div[1]/div")          
for private in pvtt:
    try:
        pvt.append(private.text)
    except NoSuchElementException:    
        pvt.append("-")
            
            
dormy=driver.find_elements(By.XPATH, "//div[@class='prices-col']//a[1]/div[2]/div")       
for dorms in dormy:
    try:
        dorm.append(dorms.text)
    except NoSuchElementException:    
        dorm.append("-")
            
faci= driver.find_elements(By.XPATH, "//div[@class='has-wifi']//span")          
    
for f in faci:
    try:
        fac.append(f.text)
    except NoSuchElementException:    
        fac.append("-") 
        
desc=driver.find_elements(By.XPATH, "//div[@class='has-wifi']//span")

for d in desc:
    try:
        des.append(d.text)
    except NoSuchElementException:
        des.append("-")
```
